# 1.DataSet Processing


```python
import json

dev_claims_baseline = json.load(open("dev-claims-baseline.json"))
dev_claims = json.load(open("dev-claims.json"))
evidence = json.load(open("evidence.json"))
test_claims_unlabelled = json.load(open("test-claims-unlabelled.json"))
train_claims = json.load(open("train-claims.json"))

# print(train_claims)
```

```python
def add_evidence_text(claims, evidence):
  dataset = {}
  for claim_key, claim_value in claims.items():
    claim_text = claim_value["claim_text"]
    claim_label = claim_value["claim_label"]
    evidences_id = claim_value.get("evidences",[])

    evidence_texts = []
    for evidence_id in evidences_id:
      if evidence_id in evidence:
        evidence_texts.append(evidence[evidence_id])

    dataset[claim_key] = {
            "claim_text": claim_text,
            "evidence_text": " ".join(evidence_texts),
            "claim_label": claim_label
    }

    return dataset


train_claims = add_evidence_text(train_claims, evidence)
dev_claims = add_evidence_text(dev_claims, evidence)
```
